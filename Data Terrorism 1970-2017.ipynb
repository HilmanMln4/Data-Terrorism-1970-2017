```python
# Punya Hilman Maulana 1706985975 - Tugas UAS Probstok 04 Pak Alfan
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import folium
from folium.plugins import HeatMap
from folium import plugins
import seaborn as sns
import plotly.graph_objs as go 
%matplotlib inline
```

```python
data = pd.read_csv('global_terrorism1970-2017.csv')
```

```python
data
```

```python
data1 = data.fillna(0)
data1.head()
```

```python
# multiple markers using dataframe
# there is an example below using apply function instead of loop

# create map
map_terrorism = folium.Map(location=[38, -98], zoom_start=1)

# plot airport locations
for (index, row) in data1.iterrows():
    folium.Marker(location=[row.loc['latitude'], row.loc['longitude']], 
                  popup=row.loc['attacktype1_txt'] + ' ' + row.loc['targtype1_txt'], 
                  tooltip='click').add_to(map_terrorism)
    
# display map    
map_terrorism
```

```python
Jumlah_nkill = data1.groupby(['region_txt', 'iyear'])['nkill'].agg(['sum']).reset_index()
Jumlah_nkill
```

```python
plt.plot(Jumlah_nkill.iyear, Jumlah_nkill.region_txt)
plt.show()
```

```python
Jumlah_pertahun = data1.groupby('iyear')['nkill'].agg(['sum']).reset_index()
Jumlah_pertahun.head()
```
